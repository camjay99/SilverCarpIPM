$$\frac{dL}{dt} = -K(L - (L_{\infty L} - gB_t))$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters:
L_0 = 10
K = 0.26
L_inf = 332
g = 0.2
def B_t(t):
    return 100*(t**(1/2))

x_grid = np.linspace(0, 100, 101)
y_grid = B_t(x_grid)
plt.plot(x_grid, y_grid)
```

```python
# Numerical Solving (Euler's Method)
n = 10
def integrate(t, L):
    h = 1/n
    for x in range(n):
        dL = h*-K*(L - (L_inf - g*B_t(t)))
        L += dL
    return L
lengths = [L_0]
for t in range(20):
    lengths.append(integrate(t, lengths[-1]))

t_grid = range(21)
plt.plot(t_grid, lengths)
plt.title('Numerical Integration', fontsize=20)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Total Length', fontsize=15)
plt.show()
```

$$L(t+1) = (L_{\infty L} - gB_t)(1 - e^{-K}) + e^{-K}L$$

```python
import math

# Substitution Formula
lengths_2 = [L_0]
for t in range(20):
    lengths_2.append((L_inf - g*B_t(t))*(1 - math.exp(-K)) + math.exp(-K)*lengths_2[-1])
    
t_grid = range(21)
plt.plot(t_grid, lengths_2)
plt.title('Substitution', fontsize=20)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Total Length', fontsize=15)
plt.show()
```

```python
t_grid = range(21)
plt.plot(t_grid, lengths, label='Numerical')
plt.plot(t_grid, lengths_2, label='Subsitution')
plt.legend()
plt.show()
```

$$\frac{dL}{dt} = -(K - hB_t)(L - L_\infty)$$

```python
h = 0.00001

# Numerical Solving (Euler's Method)
n = 10
def integrate(t, L):
    s = 1/n
    for x in range(n):
        dL = s*-(K*np.exp(-h*B_t(t)))*(L - L_inf)
        L += dL
    return L
lengths = [L_0]
for t in range(20):
    lengths.append(integrate(t, lengths[-1]))

t_grid = range(21)
plt.plot(t_grid, lengths)
plt.show()
```

```python
# Substitution Formula
lengths_2 = [L_0]
for t in range(20):
    lengths_2.append(L_inf*(1 - math.exp(-(K*np.exp(-h*B_t(t))))) + math.exp(-(K*np.exp(-h*B_t(t))))*lengths_2[-1])
    
t_grid = range(21)
plt.plot(t_grid, lengths_2)
plt.show()
```

```python
t_grid = range(21)
plt.plot(t_grid, lengths, label='Numerical')
plt.plot(t_grid, lengths_2, label='Subsitution')
plt.legend()
plt.show()
```
